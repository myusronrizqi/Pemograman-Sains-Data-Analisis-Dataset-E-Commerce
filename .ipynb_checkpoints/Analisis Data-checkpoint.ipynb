# 2. GATHERING DATA

```python
def gather_data(folder_path):
    """
    Fungsi untuk membaca semua file CSV dalam folder dan menampilkan preview datanya.
    """
    # Melakukan iterasi untuk setiap file yang ada di dalam folder_path
    for filename in os.listdir(folder_path):  
        # Memeriksa apakah file tersebut memiliki ekstensi .csv
        if filename.endswith('.csv'):  
            # Menggabungkan path folder dengan nama file untuk mendapatkan path lengkap file
            file_path = os.path.join(folder_path, filename)  
            # Membaca file CSV menggunakan pandas dan menyimpannya dalam DataFrame (df)
            df = pd.read_csv(file_path)  
            # Menampilkan nama file yang sedang diproses
            print(f"File: {filename}")  
            # Menampilkan 5 baris pertama dari DataFrame sebagai preview data
            print(df.head())  
            # Menampilkan garis pemisah antara file satu dengan file berikutnya
            print("\n" + "="*40 + "\n")  

# Path folder yang berisi dataset CSV
folder_path = 'E-Commerce Public Dataset'
# Memanggil fungsi gather_data dengan parameter folder_path
gather_data(folder_path)
```

# 3. ASSESSING DATA

```python
def assess_all_files_combined(folder_path):
    """
    Fungsi untuk menilai data gabungan dari semua file CSV dalam folder.
    """
    # Membuat list kosong untuk menyimpan DataFrame dari setiap file CSV
    all_dfs = []  
    for filename in os.listdir(folder_path):  # Iterasi untuk setiap file di folder
        if filename.endswith('.csv'):  # Memeriksa apakah file memiliki ekstensi .csv
            file_path = os.path.join(folder_path, filename)  # Menggabungkan path folder dan nama file
            df = pd.read_csv(file_path)  # Membaca file CSV menjadi DataFrame
            all_dfs.append(df)  # Menambahkan DataFrame ke dalam list all_dfs
    # Menggabungkan semua DataFrame dalam list menjadi satu DataFrame gabungan
    combined_df = pd.concat(all_dfs, ignore_index=True)  
    print("=== PENILAIAN DATA UNTUK SEMUA FILE CSV SECARA KESELURUHAN ===\n")
    
    # Informasi dasar dataset gabungan
    print("1. Informasi Dasar Dataset Gabungan:")
    print(combined_df.info())  # Menampilkan informasi struktur dataset (kolom, tipe data, jumlah non-null)
    print("\n" + "-"*40 + "\n")  # Garis pemisah
    
    # Statistik deskriptif untuk kolom numerik
    print("2. Statistik Deskriptif:")
    print(combined_df.describe())  # Menampilkan statistik deskriptif seperti mean, min, max, dll.
    print("\n" + "-"*40 + "\n")  # Garis pemisah
    
    # Memeriksa missing values
    print("3. Missing Values per Kolom:")
    print(combined_df.isnull().sum())  # Menghitung jumlah nilai yang hilang (NaN) untuk setiap kolom
    print("\n" + "-"*40 + "\n")  # Garis pemisah
    
    # Memeriksa duplikasi data
    print("4. Jumlah Baris Duplikat:")
    print(f"Jumlah baris duplikat: {combined_df.duplicated().sum()}")  # Menghitung jumlah baris yang duplikat
    print("\n" + "-"*40 + "\n")  # Garis pemisah
    
    # Distribusi nilai unik untuk kolom kategorikal
    print("5. Distribusi Nilai Unik untuk Kolom Kategorikal:")
    for column in combined_df.select_dtypes(include='object').columns:  # Memilih kolom dengan tipe data kategorikal (string)
        print(f"Distribusi nilai unik untuk kolom '{column}':")
        print(combined_df[column].value_counts())  # Menampilkan distribusi nilai unik untuk setiap kolom kategorikal
        print("\n")
    print("\n" + "-"*40 + "\n")  # Garis pemisah
    
    # Korelasi antar variabel numerik
    numeric_columns = combined_df.select_dtypes(include='number').columns  # Memilih kolom dengan tipe data numerik
    if len(numeric_columns) > 1:  # Memastikan ada lebih dari satu kolom numerik untuk menghitung korelasi
        print("6. Matriks Korelasi Antar Variabel Numerik:")
        correlation_matrix = combined_df[numeric_columns].corr()  # Menghitung matriks korelasi
        print(correlation_matrix)  # Menampilkan matriks korelasi
        print("\n" + "-"*40 + "\n")  # Garis pemisah
    else:
        print("6. Tidak ada cukup kolom numerik untuk menghitung korelasi.")  # Pesan jika tidak ada cukup kolom numerik
        print("\n" + "-"*40 + "\n")  # Garis pemisah

# Jalankan fungsi untuk folder tertentu
assess_all_files_combined(folder_path)
```

# 4. CLEANING DATA

```python
def clean_and_display(folder_path):
    """
    Fungsi untuk membersihkan data dan menyimpan hasilnya dalam folder baru.
    """
    cleaning_results = []  # List untuk menyimpan hasil cleaning setiap file
    cleaned_folder_path = os.path.join(folder_path, 'Cleaned Dataset')  # Path folder untuk menyimpan dataset yang sudah dibersihkan
    os.makedirs(cleaned_folder_path, exist_ok=True)  # Membuat folder baru jika belum ada
    for filename in os.listdir(folder_path):  # Iterasi untuk setiap file di folder
        if filename.endswith('.csv'):  # Memeriksa apakah file memiliki ekstensi .csv
            file_path = os.path.join(folder_path, filename)  # Menggabungkan path folder dan nama file
            df = pd.read_csv(file_path)  # Membaca file CSV menjadi DataFrame
            print(f"\n=== MEMBERSIHKAN DATA UNTUK FILE: {filename} ===\n")
            
            # **Menghapus baris duplikat**
            initial_rows = df.shape[0]  # Jumlah baris awal sebelum penghapusan duplikat
            df = df.drop_duplicates()  # Menghapus baris duplikat
            duplicate_rows_removed = initial_rows - df.shape[0]  # Menghitung jumlah baris duplikat yang dihapus
            
            # **Mengisi missing values**
            missing_values_before = df.isnull().sum().to_dict()  # Menghitung jumlah missing values sebelum cleaning
            for column in df.select_dtypes(include='number').columns:  # Untuk kolom numerik
                df[column] = df[column].fillna(df[column].mean())  # Mengisi missing values dengan nilai rata-rata (mean)
            for column in df.select_dtypes(include='object').columns:  # Untuk kolom kategorikal
                df[column] = df[column].fillna(df[column].mode()[0])  # Mengisi missing values dengan nilai modus (mode)
            missing_values_after = df.isnull().sum().to_dict()  # Menghitung jumlah missing values setelah cleaning
            
            # **Memperbaiki tipe data**
            if 'date' in df.columns:  # Jika ada kolom bernama 'date'
                df['date'] = pd.to_datetime(df['date'], errors='coerce')  # Mengubah kolom 'date' menjadi tipe datetime
            for column in df.select_dtypes(include='number').columns:  # Untuk kolom numerik
                df[column] = pd.to_numeric(df[column], errors='coerce')  # Memastikan kolom numerik memiliki tipe data numerik
            
            # **Menghapus outlier menggunakan metode IQR**
            outlier_rows_removed = 0  # Inisialisasi jumlah baris outlier yang dihapus
            for column in df.select_dtypes(include='number').columns:  # Untuk kolom numerik
                Q1 = df[column].quantile(0.25)  # Menghitung kuartil pertama (Q1)
                Q3 = df[column].quantile(0.75)  # Menghitung kuartil ketiga (Q3)
                IQR = Q3 - Q1  # Menghitung Interquartile Range (IQR)
                lower_bound = Q1 - 1.5 * IQR  # Batas bawah untuk mendeteksi outlier
                upper_bound = Q3 + 1.5 * IQR  # Batas atas untuk mendeteksi outlier
                initial_rows = df.shape[0]  # Jumlah baris sebelum menghapus outlier
                df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]  # Menghapus baris outlier
                outlier_rows_removed += initial_rows - df.shape[0]  # Menghitung jumlah baris outlier yang dihapus
            
            # **Simpan hasil cleaning untuk file ini**
            result = {
                "filename": filename,  # Nama file
                "initial_rows": initial_rows,  # Jumlah baris awal
                "final_rows": df.shape[0],  # Jumlah baris akhir setelah cleaning
                "duplicate_rows_removed": duplicate_rows_removed,  # Jumlah baris duplikat yang dihapus
                "missing_values_before": missing_values_before,  # Missing values sebelum cleaning
                "missing_values_after": missing_values_after,  # Missing values setelah cleaning
                "outlier_rows_removed": outlier_rows_removed  # Jumlah baris outlier yang dihapus
            }
            cleaning_results.append(result)  # Menyimpan hasil cleaning ke dalam list
            
            # **Tampilkan hasil cleaning untuk file ini**
            print("Hasil Cleaning:")
            print(f"- Baris awal: {initial_rows}")
            print(f"- Baris akhir: {df.shape[0]}")
            print(f"- Baris duplikat yang dihapus: {duplicate_rows_removed}")
            print(f"- Missing values sebelum cleaning: {missing_values_before}")
            print(f"- Missing values setelah cleaning: {missing_values_after}")
            print(f"- Baris outlier yang dihapus: {outlier_rows_removed}")
            print("\n" + "-"*40 + "\n")  # Garis pemisah
            
            # **Simpan dataset yang sudah dibersihkan ke file CSV baru**
            cleaned_file_name = f"cleaned_{filename}"  # Nama file baru dengan awalan 'cleaned_'
            cleaned_file_path = os.path.join(cleaned_folder_path, cleaned_file_name)  # Path lengkap file baru
            df.to_csv(cleaned_file_path, index=False)  # Menyimpan DataFrame ke file CSV tanpa indeks
            print(f"Dataset yang sudah dibersihkan disimpan di: {cleaned_file_path}")
    
    # **Mengubah hasil cleaning menjadi DataFrame**
    results_df = pd.DataFrame(cleaning_results)  # Mengonversi list hasil cleaning menjadi DataFrame
    return results_df  # Mengembalikan DataFrame ringkasan hasil cleaning

# Jalankan fungsi untuk folder tertentu
cleaning_summary = clean_and_display(folder_path)

# Menampilkan ringkasan hasil cleaning
print("\n=== RINGKASAN HASIL CLEANING UNTUK SEMUA FILE ===\n")
print(cleaning_summary)
```

# 5. EDA (Exploratory Data Analysis)

```python
def perform_eda(folder_path):
    """
    Fungsi untuk melakukan Exploratory Data Analysis (EDA) pada dataset yang sudah dibersihkan.
    """
    for filename in os.listdir(folder_path):
        if filename.endswith('.csv'):  
            file_path = os.path.join(folder_path, filename)  

            try:
                df = pd.read_csv(file_path)  
                print(f"\n=== EDA UNTUK FILE: {filename} ===\n")

                # Informasi dasar dataset
                print("Informasi Dasar Dataset:")
                print(df.info())  
                print("\nStatistik Deskriptif:")
                print(df.describe())  

                # Analisis Missing Values
                print("\nAnalisis Missing Values:")
                missing_values = df.isnull().sum()
                print(missing_values[missing_values > 0])  

                # Visualisasi distribusi data numerik
                numeric_columns = df.select_dtypes(include='number').columns  
                if len(numeric_columns) > 0:
                    print("\nVisualisasi Distribusi Data Numerik:")
                    for column in numeric_columns:
                        plt.figure(figsize=(8, 4))
                        sns.histplot(df[column], kde=True)  
                        plt.title(f"Distribusi {column}")
                        plt.show()

                # Visualisasi distribusi data kategorikal
                categorical_columns = df.select_dtypes(include='object').columns  
                if len(categorical_columns) > 0:
                    print("\nVisualisasi Distribusi Data Kategorikal:")
                    for column in categorical_columns:
                        if df[column].nunique() > 20:
                            print(f"Kolom '{column}' memiliki terlalu banyak nilai unik ({df[column].nunique()}), dilewati.")
                            continue
                        plt.figure(figsize=(8, 4))
                        sns.countplot(data=df, x=column)
                        plt.title(f"Distribusi {column}")
                        plt.xticks(rotation=45)
                        plt.show()

                # Analisis korelasi antar variabel numerik
                if len(numeric_columns) > 1:
                    print("\nAnalisis Hubungan Antar Variabel Numerik:")
                    correlation_matrix = df[numeric_columns].corr()
                    print("Matriks Korelasi:")
                    print(correlation_matrix)

                    plt.figure(figsize=(10, 8))
                    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
                    plt.title("Heatmap Korelasi")
                    plt.show()

                # Deteksi outlier
                if len(numeric_columns) > 0:
                    print("\nDeteksi Outlier:")
                    for column in numeric_columns:
                        plt.figure(figsize=(8, 4))
                        sns.boxplot(data=df, x=column)
                        plt.title(f"Boxplot untuk {column}")
                        plt.show()

                # Analisis time series
                date_columns = [col for col in df.columns if 'date' in col.lower() or 'timestamp' in col.lower()]
                if len(date_columns) > 0:
                    print("\nAnalisis Waktu:")
                    for column in date_columns:
                        df[column] = pd.to_datetime(df[column], errors='coerce')
                        if df[column].isnull().all():
                            print(f"Kolom '{column}' tidak dapat dikonversi ke datetime, dilewati.")
                            continue

                        df['year'] = df[column].dt.year
                        df['month'] = df[column].dt.month

                        yearly_counts = df.groupby('year').size()
                        plt.figure(figsize=(10, 6))
                        yearly_counts.plot(kind='line', title=f"Tren Jumlah Data Berdasarkan Tahun ({column})")
                        plt.show()

                # Analisis per grup
                if len(categorical_columns) > 0 and len(numeric_columns) > 0:
                    print("\nAnalisis Berdasarkan Grup:")
                    for cat_col in categorical_columns:
                        if df[cat_col].nunique() > 10:
                            print(f"Kolom '{cat_col}' memiliki terlalu banyak nilai unik ({df[cat_col].nunique()}), dilewati.")
                            continue
                        for num_col in numeric_columns:
                            plt.figure(figsize=(8, 4))
                            sns.boxplot(data=df, x=cat_col, y=num_col)
                            plt.title(f"Distribusi {num_col} Berdasarkan {cat_col}")
                            plt.xticks(rotation=45)
                            plt.show()

                print("\n" + "-"*40 + "\n")

            except Exception as e:
                print(f"Gagal memproses file {filename}: {e}")

# Menjalankan EDA untuk folder dataset yang sudah dibersihkan
perform_eda('E-Commerce Public Dataset/Cleaned Dataset')
```

# 6. FEATURE ENGINEERING

```python
# Muat dataset yang sudah dibersihkan
sellers_file_path = 'E-Commerce Public Dataset/Cleaned Dataset/cleaned_sellers_dataset.csv'
order_items_file_path = 'E-Commerce Public Dataset/Cleaned Dataset/cleaned_order_items_dataset.csv'
orders_file_path = 'E-Commerce Public Dataset/Cleaned Dataset/cleaned_orders_dataset.csv'
customers_file_path = 'E-Commerce Public Dataset/Cleaned Dataset/cleaned_customers_dataset.csv'
geolocation_file_path = 'E-Commerce Public Dataset/Cleaned Dataset/cleaned_geolocation_dataset.csv'

sellers_df = pd.read_csv(sellers_file_path)
order_items_df = pd.read_csv(order_items_file_path)
orders_df = pd.read_csv(orders_file_path)
customers_df = pd.read_csv(customers_file_path)
geolocation_df = pd.read_csv(geolocation_file_path)

# 1. Hitung total pesanan, harga rata-rata, dan total pendapatan per penjual
order_counts = order_items_df.groupby('seller_id').agg(
    total_orders=('order_id', 'nunique'),  # Total pesanan
    average_price=('price', 'mean'),      # Rata-rata harga produk
    total_revenue=('price', 'sum')         # Total pendapatan
).reset_index()

# Hitung total pendapatan termasuk nilai pengiriman
order_counts['total_revenue'] = order_counts['total_revenue'] + order_items_df.groupby('seller_id')['freight_value'].sum().reset_index()['freight_value']

# Gabungkan dengan sellers_df
sellers_df = sellers_df.merge(order_counts, on='seller_id', how='left')

# 2. Hitung waktu pengiriman rata-rata per penjual
# Gabungkan order_items_df dengan orders_df untuk mendapatkan tanggal pengiriman
order_items_orders_df = order_items_df.merge(orders_df, on='order_id', how='left')

# Ubah tanggal menjadi datetime
order_items_orders_df['order_delivered_customer_date'] = pd.to_datetime(order_items_orders_df['order_delivered_customer_date'], errors='coerce')
order_items_orders_df['order_purchase_timestamp'] = pd.to_datetime(order_items_orders_df['order_purchase_timestamp'], errors='coerce')

# Hitung waktu pengiriman dalam hitungan hari
order_items_orders_df['delivery_time_days'] = (order_items_orders_df['order_delivered_customer_date'] - order_items_orders_df['order_purchase_timestamp']).dt.days

# Hitung waktu pengiriman rata-rata per penjual
average_delivery_time = order_items_orders_df.groupby('seller_id')['delivery_time_days'].mean().reset_index()
average_delivery_time.rename(columns={'delivery_time_days': 'average_delivery_time'}, inplace=True)

# Gabungkan dengan sellers_df
sellers_df = sellers_df.merge(average_delivery_time, on='seller_id', how='left')

# 3. Hitung frekuensi transaksi per bulan/tahun untuk setiap penjual
order_items_orders_df['order_month'] = order_items_orders_df['order_purchase_timestamp'].dt.to_period('M')
order_items_orders_df['order_year'] = order_items_orders_df['order_purchase_timestamp'].dt.year

transaction_frequency = order_items_orders_df.groupby(['seller_id', 'order_month']).agg({
    'order_id': 'nunique'
}).reset_index()

transaction_frequency = transaction_frequency.groupby('seller_id').agg({
    'order_id': 'mean'
}).reset_index()

transaction_frequency.rename(columns={'order_id': 'average_monthly_transactions'}, inplace=True)

# Gabungkan dengan sellers_df
sellers_df = sellers_df.merge(transaction_frequency, on='seller_id', how='left')

# 4. Hitung jarak rata-rata antara penjual dan pelanggan
# Petakan data geolokasi untuk penjual
seller_geo_mapping = geolocation_df.groupby('geolocation_zip_code_prefix').agg({
    'geolocation_lat': 'mean',
    'geolocation_lng': 'mean'
}).reset_index()

# Petakan data geolokasi untuk pelanggan
customer_geo_mapping = geolocation_df.groupby('geolocation_zip_code_prefix').agg({
    'geolocation_lat': 'mean',
    'geolocation_lng': 'mean'
}).reset_index()

# Gabungkan sellers_df dengan order_items_orders_df untuk mendapatkan ID pelanggan
order_items_orders_sellers_customers_df = order_items_orders_df.merge(sellers_df, on='seller_id', how='left').merge(customers_df, on='customer_id', how='left')

# Gabungkan dengan data geolokasi untuk penjual
order_items_orders_sellers_customers_df = order_items_orders_sellers_customers_df.merge(
    seller_geo_mapping, left_on='seller_zip_code_prefix', right_on='geolocation_zip_code_prefix', how='left'
)

# Gabungkan dengan data geolokasi untuk pelanggan
order_items_orders_sellers_customers_df = order_items_orders_sellers_customers_df.merge(
    customer_geo_mapping, left_on='customer_zip_code_prefix', right_on='geolocation_zip_code_prefix', how='left', suffixes=('_seller', '_customer')
)

# Hitung jarak rata-rata antara penjual dan pelanggan
order_items_orders_sellers_customers_df['distance'] = order_items_orders_sellers_customers_df.apply(
    lambda row: haversine((row['geolocation_lat_seller'], row['geolocation_lng_seller']), 
                          (row['geolocation_lat_customer'], row['geolocation_lng_customer'])), axis=1
)

average_distance = order_items_orders_sellers_customers_df.groupby('seller_id')['distance'].mean().reset_index()
average_distance.rename(columns={'distance': 'average_distance_to_customers'}, inplace=True)

# Gabungkan dengan seller_df
sellers_df = sellers_df.merge(average_distance, on='seller_id', how='left')

# 5. Fitur Regional - Jumlah penjual di setiap negara bagian atau kota
seller_count_by_city = sellers_df.groupby('seller_city').size().reset_index(name='seller_count_by_city')
sellers_df = sellers_df.merge(seller_count_by_city, on='seller_city', how='left')

seller_count_by_state = sellers_df.groupby('seller_state').size().reset_index(name='seller_count_by_state')
sellers_df = sellers_df.merge(seller_count_by_state, on='seller_state', how='left')

# 6. Isi nilai yang hilang untuk fitur baru
sellers_df['total_orders'] = sellers_df['total_orders'].fillna(0)
sellers_df['average_price'] = sellers_df['average_price'].fillna(0)
sellers_df['total_revenue'] = sellers_df['total_revenue'].fillna(0)
sellers_df['average_delivery_time'] = sellers_df['average_delivery_time'].fillna(0)
sellers_df['average_monthly_transactions'] = sellers_df['average_monthly_transactions'].fillna(0)
sellers_df['average_distance_to_customers'] = sellers_df['average_distance_to_customers'].fillna(0)
sellers_df['seller_count_by_city'] = sellers_df['seller_count_by_city'].fillna(0)
sellers_df['seller_count_by_state'] = sellers_df['seller_count_by_state'].fillna(0)

# 7. Tampilkan kumpulan data yang diperbarui dengan fitur-fitur baru
print(sellers_df.head())

# 8. Buat folder Feature Engineering jika belum ada
feature_engineered_folder = 'E-Commerce Public Dataset/Feature Engineered Dataset'
os.makedirs(feature_engineered_folder, exist_ok=True)

# 9. Simpan kumpulan data yang diperbarui ke file baru di folder Feature Engineering
output_file_path = os.path.join(feature_engineered_folder, 'feature_engineered_cleaned_sellers_dataset.csv')
sellers_df.to_csv(output_file_path, index=False)
print(f"Dataset dengan fitur baru telah disimpan di: {output_file_path}")
```

# 7. PEMBAGIAN DATASET & DATA MINING (Regresi dan Klasifikasi)

## Model Regresi untuk prediksi total pendapatan (total_revenue)

```python
# Muat dataset yang telah dilakukan feature engineering
feature_engineered_file_path = 'E-Commerce Public Dataset/Feature Engineered Dataset/feature_engineered_cleaned_sellers_dataset.csv'
sellers_df = pd.read_csv(feature_engineered_file_path)

# Langkah 1: Tentukan fitur (features) dan label (target)
# Features: Fitur numerik yang relevan
X_regression = sellers_df[['total_orders', 'average_price', 'average_monthly_transactions', 
                           'average_distance_to_customers', 'seller_count_by_city', 'seller_count_by_state']]
# Target: Variabel yang ingin diprediksi (total_revenue)
y_regression = sellers_df['total_revenue']

# Langkah 2: Pisahkan data menjadi data pelatihan dan pengujian
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_regression, y_regression, test_size=0.2, random_state=42)

# Langkah 3: Normalisasi data
scaler = StandardScaler()
X_train_reg_scaled = scaler.fit_transform(X_train_reg)
X_test_reg_scaled = scaler.transform(X_test_reg)

# Langkah 4: Latih model regresi menggunakan Random Forest
# Definisikan Random Forest Regressor
rf_regressor = RandomForestRegressor(random_state=42)

# Hyperparameter tuning menggunakan GridSearchCV
param_grid = {
    'n_estimators': [100, 200, 300],  # Jumlah pohon dalam forest
    'max_depth': [None, 10, 20, 30],  # Kedalaman maksimum pohon
    'min_samples_split': [2, 5, 10],  # Minimum jumlah sampel untuk split
    'min_samples_leaf': [1, 2, 4]     # Minimum jumlah sampel di leaf node
}

# GridSearchCV untuk mencari kombinasi hyperparameter terbaik
grid_search = GridSearchCV(estimator=rf_regressor, param_grid=param_grid, 
                           cv=3, scoring='neg_mean_squared_error', n_jobs=-1, verbose=2)
grid_search.fit(X_train_reg_scaled, y_train_reg)

# Model terbaik dari GridSearchCV
best_rf_model = grid_search.best_estimator_

# Langkah 5: Evaluasi model pada data pengujian
y_pred_reg = best_rf_model.predict(X_test_reg_scaled)

# Hitung metrik evaluasi
mse = mean_squared_error(y_test_reg, y_pred_reg)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test_reg, y_pred_reg)
r2 = r2_score(y_test_reg, y_pred_reg)

# Tampilkan hasil evaluasi
print("Hasil Evaluasi Model Regresi:")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"R-squared (R²): {r2:.2f}")

# Tampilkan hyperparameter terbaik
print("\nHyperparameter Terbaik:")
print(grid_search.best_params_)

# Langkah 6: Simpan hasil prediksi ke dalam DataFrame untuk analisis lebih lanjut
results_df = pd.DataFrame({
    'Actual Revenue': y_test_reg,
    'Predicted Revenue': y_pred_reg
})
print("\nHasil Prediksi:")
print(results_df.head())

# Langkah 7: Simpan model
model_directory = 'E-Commerce Public Dataset/Models'
os.makedirs(model_directory, exist_ok=True)  # Buat direktori jika belum ada

# Simpan model menggunakan joblib
model_file_path = os.path.join(model_directory, 'random_forest_regression_model.pkl')
joblib.dump(best_rf_model, model_file_path)
print(f"\nModel regresi telah disimpan di: {model_file_path}")

# Membuat Visualisasi
# 1. Scatter Plot: Actual vs Predicted
plt.figure(figsize=(8, 6))
plt.scatter(results_df['Actual Revenue'], results_df['Predicted Revenue'], alpha=0.7)
plt.plot([results_df['Actual Revenue'].min(), results_df['Actual Revenue'].max()],
         [results_df['Actual Revenue'].min(), results_df['Actual Revenue'].max()], 'r--', label='Ideal Fit')
plt.xlabel('Pendapatan Aktual')
plt.ylabel('Pendapatan Prediksi')
plt.title('Pendapatan Aktual vs Pendapatan Prediksi')
plt.legend()
plt.show()

# 2. Residual Plot
residuals = results_df['Actual Revenue'] - results_df['Predicted Revenue']
plt.figure(figsize=(8, 6))
plt.scatter(results_df['Predicted Revenue'], residuals, alpha=0.7)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('Pendapatan Prediksi')
plt.ylabel('Residual')
plt.title('Plot Residual')
plt.show()

# 3. Histogram Residual
plt.figure(figsize=(8, 6))
sns.histplot(residuals, kde=True, bins=30, color='blue')
plt.axvline(0, color='red', linestyle='--')
plt.xlabel('Residual')
plt.ylabel('Frekuensi')
plt.title('Histogram Residual')
plt.show()

# 4. Feature Importance
importances = best_rf_model.feature_importances_
feature_names = X_regression.columns
feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(data=feature_importance_df, x='Importance', y='Feature', palette='viridis')
plt.title('Kepentingan Fitur')
plt.xlabel('Skor Kepentingan')
plt.ylabel('Fitur')
plt.show()
```

## Model Klasifikasi untuk prediksi penjual aktif atau tidak aktif (activity_status)

```python
# Muat dataset yang telah dilakukan feature engineering
feature_engineered_file_path = 'E-Commerce Public Dataset/Feature Engineered Dataset/feature_engineered_cleaned_sellers_dataset.csv'
sellers_df = pd.read_csv(feature_engineered_file_path)  # Membaca file CSV ke dalam DataFrame

# Langkah 1: Tentukan fitur (features) dan label (target)
# Features: Fitur numerik yang relevan untuk digunakan dalam model
X_classification = sellers_df[['total_orders', 'average_price', 'average_monthly_transactions', 
                               'average_distance_to_customers', 'seller_count_by_city', 'seller_count_by_state']]
# Target: Buat label baru untuk klasifikasi (active atau not active)
def categorize_activity(total_orders):
    return 'Active' if total_orders > 0 else 'Not Active'  # Jika total_orders > 0, maka status "Active", jika tidak "Not Active"
sellers_df['activity_status'] = sellers_df['total_orders'].apply(categorize_activity)  # Menambahkan kolom baru berdasarkan fungsi
y_classification = sellers_df['activity_status']  # Label target untuk klasifikasi

# Langkah 2: Ubah label kategorikal menjadi numerik menggunakan LabelEncoder
label_encoder = LabelEncoder()
y_classification_encoded = label_encoder.fit_transform(y_classification)  # Mengubah label kategorikal menjadi numerik

# Langkah 3: Pisahkan data menjadi data pelatihan dan pengujian
# Menggunakan train_test_split untuk membagi data menjadi 80% pelatihan dan 20% pengujian
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(X_classification, y_classification_encoded, test_size=0.2, random_state=42)

# Langkah 4: Normalisasi data
# Menggunakan StandardScaler untuk menormalisasi fitur agar memiliki mean 0 dan variance 1
scaler = StandardScaler()
X_train_clf_scaled = scaler.fit_transform(X_train_clf)  # Normalisasi data pelatihan
X_test_clf_scaled = scaler.transform(X_test_clf)       # Normalisasi data pengujian

# Langkah 5: Latih model klasifikasi menggunakan XGBoost
# Definisikan XGBoost Classifier sebagai model klasifikasi
xgb_classifier = XGBClassifier(random_state=42, eval_metric='logloss')

# Hyperparameter tuning menggunakan GridSearchCV
# Menentukan parameter yang akan diuji untuk mencari kombinasi terbaik
param_grid = {
    'n_estimators': [100, 200],          # Jumlah pohon dalam model
    'max_depth': [3, 5, 7],             # Kedalaman maksimum setiap pohon
    'learning_rate': [0.01, 0.1, 0.2],  # Tingkat pembelajaran
    'subsample': [0.8, 0.9],            # Proporsi sampel yang digunakan untuk setiap pohon
    'colsample_bytree': [0.8, 0.9]      # Proporsi fitur yang digunakan untuk setiap pohon
}

# GridSearchCV untuk mencari kombinasi hyperparameter terbaik
# cv=3 berarti menggunakan 3-fold cross-validation
grid_search = GridSearchCV(estimator=xgb_classifier, param_grid=param_grid, 
                           cv=3, scoring='accuracy', n_jobs=-1, verbose=2)
grid_search.fit(X_train_clf_scaled, y_train_clf)  # Melatih model dengan data pelatihan

# Model terbaik dari GridSearchCV
best_xgb_model = grid_search.best_estimator_  # Mendapatkan model dengan kombinasi hyperparameter terbaik

# Langkah 6: Evaluasi model pada data pengujian
y_pred_clf = best_xgb_model.predict(X_test_clf_scaled)  # Melakukan prediksi pada data pengujian

# Ubah prediksi numerik kembali ke label kategorikal
y_test_clf_labels = label_encoder.inverse_transform(y_test_clf)  # Mengembalikan label numerik ke kategorikal
y_pred_clf_labels = label_encoder.inverse_transform(y_pred_clf)  # Mengembalikan prediksi numerik ke kategorikal

# Tampilkan hasil evaluasi
print("Hasil Evaluasi Model Klasifikasi:")
print(f"Accuracy Score: {accuracy_score(y_test_clf, y_pred_clf):.2f}")  # Menampilkan akurasi
print("\nClassification Report:")
print(classification_report(y_test_clf_labels, y_pred_clf_labels))  # Menampilkan classification report
print("\nConfusion Matrix:")
print(confusion_matrix(y_test_clf_labels, y_pred_clf_labels))  # Menampilkan confusion matrix

# Tampilkan hyperparameter terbaik
print("\nHyperparameter Terbaik:")
print(grid_search.best_params_)  # Menampilkan kombinasi hyperparameter terbaik

# Langkah 7: Simpan hasil prediksi ke dalam DataFrame untuk analisis lebih lanjut
results_df = pd.DataFrame({
    'Actual Activity Status': y_test_clf_labels,  # Status aktivitas aktual
    'Predicted Activity Status': y_pred_clf_labels  # Status aktivitas hasil prediksi
})
print("\nHasil Prediksi:")
print(results_df.head())  # Menampilkan 5 baris pertama hasil prediksi

# Langkah 8: Simpan model 
model_directory = 'E-Commerce Public Dataset/Models'
os.makedirs(model_directory, exist_ok=True)  # Membuat direktori jika belum ada

# Simpan model menggunakan joblib
model_file_path = os.path.join(model_directory, 'xgboost_classification_model.pkl')
joblib.dump(best_xgb_model, model_file_path)  # Menyimpan model ke file .pkl
print(f"\nModel klasifikasi telah disimpan di: {model_file_path}")

# Membuat Visualisasi
# 1. Confusion Matrix Heatmap
conf_matrix = confusion_matrix(y_test_clf_labels, y_pred_clf_labels)  # Menghitung confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Not Active', 'Active'], yticklabels=['Not Active', 'Active'])
plt.xlabel('Prediksi')  # Label sumbu x
plt.ylabel('Aktual')  # Label sumbu y
plt.title('Heatmap Confusion Matrix')  # Judul plot
plt.show()

# 2. ROC Curve dan AUC Score
y_pred_prob = best_xgb_model.predict_proba(X_test_clf_scaled)[:, 1]  # Probabilitas untuk kelas positif (Active)
fpr, tpr, thresholds = roc_curve(y_test_clf, y_pred_prob)  # Menghitung false positive rate dan true positive rate
roc_auc = auc(fpr, tpr)  # Menghitung area under curve (AUC)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'Kurva ROC (AUC = {roc_auc:.2f})', color='blue')  # Plot ROC curve
plt.plot([0, 1], [0, 1], 'r--', label='Tebakan Acak')  # Garis acak
plt.xlabel('False Positive Rate')  # Label sumbu x
plt.ylabel('True Positive Rate')  # Label sumbu y
plt.title('Kurva ROC')  # Judul plot
plt.legend()  # Menampilkan legenda
plt.show()

# 3. Precision-Recall Curve
precision, recall, _ = precision_recall_curve(y_test_clf, y_pred_prob)  # Menghitung precision dan recall
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, label='Kurva Precision-Recall', color='green')  # Plot precision-recall curve
plt.xlabel('Recall')  # Label sumbu x
plt.ylabel('Precision')  # Label sumbu y
plt.title('Kurva Precision-Recall')  # Judul plot
plt.legend()  # Menampilkan legenda
plt.show()

# 4. Feature Importance
importances = best_xgb_model.feature_importances_  # Mengambil skor kepentingan fitur
feature_names = X_classification.columns  # Nama fitur
feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})  # Buat DataFrame
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)  # Urutkan berdasarkan skor kepentingan
plt.figure(figsize=(10, 6))
sns.barplot(data=feature_importance_df, x='Importance', y='Feature', palette='viridis')  # Bar plot kepentingan fitur
plt.title('Kepentingan Fitur')  # Judul plot
plt.xlabel('Skor Kepentingan')  # Label sumbu x
plt.ylabel('Fitur')  # Label sumbu y
plt.show()
```
